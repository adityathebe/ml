# Acces rows via index

df[9] searches for column named 9. What if we want to acces the 9th row ? In that case we use *.loc[]*

```python
df.loc[9]
```

### By Default Pandas assigns the index automatically starting from 0. We can change the index by using the set_index() method

```python
df.set_index('PassengerId', inplace=True)
```

```python
df.head()
```

### We can also reset the index

```python
df.reset_index(inplace=True)
df.head(2)
```
